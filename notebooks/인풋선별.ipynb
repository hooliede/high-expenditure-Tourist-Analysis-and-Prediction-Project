```python
import pandas as pd

df = pd.read_csv("c:/Users/tj/Desktop/정규프로젝트/join_data/merged_total.csv", encoding='ISO-8859-1')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 박스 플롯 생성
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='GRADUATE', y='TOTAL_PAYMENT')  # TOTAL_PAYMENT가 올바른 열 이름인지 확인
plt.title('Distribution of TOTAL_PAYMENT by Graduate Status')
plt.xlabel('Graduate Status (0: Non-Graduate, 1: Graduate)')
plt.ylabel('Total Payment')
plt.xticks(ticks=[0, 1], labels=['Non-Graduate', 'Graduate'])
plt.yscale('log')

plt.show()
```

```python
# 박스 플롯 생성
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='GRADUATE', y='TOTAL_PAYMENT')
plt.title('Distribution of TOTAL_PAYMENT by Graduate Status')
plt.xlabel('Graduate Status (0: Non-Graduate, 1: Graduate)')
plt.ylabel('Total Payment')

# y축 범위 설정
plt.ylim(0, 1000000)  # 원하는 범위로 설정

plt.xticks(ticks=[0, 1], labels=['Non-Graduate', 'Graduate'])
plt.show()
```

```python
# 지출 수준에 대한 막대 그래프 (예시로 'PAYMENT_GRADE'가 존재한다고 가정)
payment_grade_counts = df.groupby(['GRADUATE_OR_ABOVE', 'PAYMENT_GRADE']).size().unstack()

# 막대 그래프 시각화
payment_grade_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Expenditure Level by Graduate Status')
plt.xlabel('Graduate Status (0: Non-Graduate, 1: Graduate or Above)')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Graduate', 'Graduate or Above'])
plt.legend(title='Payment Grade')
plt.show()
```
